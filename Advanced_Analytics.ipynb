## Advanced Insights

1. The fund with the highest historical VaR is the one with the largest downside tail in daily returns; this indicates the highest sensitivity to adverse market moves.
2. Investor cohorts entering in earlier years tend to contribute the greatest cumulative invested capital, suggesting stronger commitment and retention.
3. SIP continuity is strongest for investors maintaining regular contributions without large gaps; investors with gaps above 35 days should be flagged for follow-up.
4. Equity funds with higher HHI values are more concentrated and therefore carry more idiosyncratic sector exposure.
5. The recommended funds are selected by Sharpe ratio within the matching risk grade to balance reward and volatility.
